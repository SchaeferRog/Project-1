```python
####Importing Pandas and Loading the CSV, turn into DataFrame, and view the data

import pandas as pd
air_data = pd.read_csv('AirQuality_Daily.csv')

#convert to dataframe
air_data = pd.DataFrame(air_data)

#view the data
air_data.head(5)
```

```python
####Task 1 - What are the 5 locations in Nebraska with the highest mean and median concentrations of VOC, PM 2.5 and PM 10.0?

#creating a group by sensor.name and finding stat summary for mean and median
air_data_voc = air_data.groupby(['sensor.name']).agg(
                                                     mean = ('voc','mean'),
                                                     median = ('voc','median'))
air_data_pm25 = air_data.groupby(['sensor.name']).agg(
                                                     mean = ('pm2.5_atm','mean'),
                                                     median = ('pm2.5_atm','median'))
air_data_pm10 = air_data.groupby(['sensor.name']).agg(
                                                     mean = ('pm10.0_atm','mean'),
                                                     median = ('pm10.0_atm','median'))


#find the top 5 median and mean for each pollutant                                  
top5_voc_combined = air_data_voc.sort_values(
    by=['mean', 'median'],
    ascending=[False, False]
).head(5)

top5_pm25_combined = air_data_pm25.sort_values(
    by=['mean', 'median'],
    ascending=[False, False]
).head(5)

top5_pm10_combined = air_data_pm10.sort_values(
    by=['mean', 'median'],
    ascending=[False, False]
).head(5)


#print the tables
top5_voc_combined
```

```python
####Task 2 - On what days did the maximum values occur and where did this maximums occur? And potential reasons? 

#find the max value of each pollutant
voc_max = air_data.agg(max = ('voc', 'max')).iloc[0,0]
pm25_max = air_data.agg(max = ('pm2.5_atm', 'max')).iloc[0,0]
pm10_max = air_data.agg(max = ('pm10.0_atm', 'max')).iloc[0,0]

#filter the data for max, date, and locations 
voc_max_row = air_data[air_data['voc'] == voc_max]
pm25_max_row = air_data[air_data['pm2.5_atm'] == pm25_max]
pm10_max_row = air_data[air_data['pm10.0_atm'] == pm10_max]

# the maximums with date and location 
voc_max_row
```

```python
####Task 3 - Does humidity and temperature have a noticeable effect on air quality?

#Create temperature categories based on the tables 
def assign_temperature(temperature):
    if temperature < 32: 
        return 'Below Freezing'
    elif 32 <= temperature <= 50:
        return 'Cool'
    elif 50 < temperature <= 70:
        return 'Warm'
    else:
        return 'Hot' #this line means that if temperature is >70 it is Hot 

#Apply temperature function to the temperature column in air_data
air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature) 

#Create hmidity categories based on humidity levels
def assign_humidity(humidity):
    if humidity < 50: 
        return 'Low'
    elif 50 <= humidity <= 80:
        return 'High'
    else:
        return 'Very High' 

#Apply humidity function to the temperature column in air_data
air_data['humidity_level'] = air_data['humidity'].apply(assign_humidity) 
        
#Determine if temperature has an effect on air quality by making a pivot table 
mean_values_temperature = pd.pivot_table(
    air_data,
    index='temperature_level',
    values=["voc","pm2.5_atm","pm10.0_atm"],
    aggfunc=['mean']).sort_index()

#Determine if humidity has an effect on air quality by making a pivot table 
mean_values_humidity = pd.pivot_table(
    air_data,
    index='humidity_level',
    values=["voc","pm2.5_atm","pm10.0_atm"],
    aggfunc=['mean']).sort_index()


#Choose which variable to look at by running that pivot table
mean_values_temperature
```

```python
####Task 4 - AQI health risks? And when? 
#Unhealthy for sensitive groups @ pm2.5 35.5 - 55.4 
#Unhealthy for sensitibe groups @ pm10.0 155 - 254

#filter data for specific ranges
unhealthy_pm25 = air_data[(air_data['pm2.5_atm']>=35.5) & (air_data['pm2.5_atm']<=55.4)]
unhealthy_pm10 = air_data[(air_data['pm2.5_atm']>=155) & (air_data['pm2.5_atm']<=254)]

#sort by date and sensor name 
unhealthy_pm25 = unhealthy_pm25[
    ['date','sensor.name','pm2.5_atm']
    ].sort_values('date')
unhealthy_pm10 = unhealthy_pm10[
    ['date','sensor.name','pm10.0_atm']
    ].sort_values('date')

#print to see dates and location
unhealthy_pm25
```
